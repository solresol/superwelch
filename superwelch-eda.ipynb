```python
import sqlite3
import pandas
import matplotlib.pyplot
```

```python
conn = sqlite3.connect("simulation_database.sqlite")
```

```python
simulations = pandas.read_sql("select * from simulations", conn)
simulations.sample(5)
```

```python
welch = pandas.read_sql("select * from welch", conn)
welch.sample(5)
```

```python
rfc = pandas.read_sql("select * from random_forest", conn)
rfc.sample(5)
```

```python
df = simulations.merge(rfc, on='simulation_id', suffixes=('', '_rfc')
                      ).merge(welch, on='simulation_id', suffixes=('_rfc', '_welch'))
df
```

```python
df5 = df[(df.tag == 'Participant size vs training size') 
         & (df.control_group_size == 5) & (df.experiment_group_size == 5)]
df5_pivot = df5.pivot_table(index='number_of_training_experiments', aggfunc=[min, max, 'mean', 'std', len])
```

```python
df5_pivot.head(10)
```

```python
fig, ax = matplotlib.pyplot.subplots()
ax.set_xscale('log')
ax.set_xlabel("Number of training experiments")
ax.set_title("Min, max and mean accuracy over 5 generated data sets")
#ax.set_ylim((0,1))
df5_pivot[('mean', 'accuracy_rfc')].plot(ax=ax, label="Random forest classifier", color="red")
df5_pivot[('min', 'accuracy_rfc')].plot(ax=ax, label="Random forest classifier", color="red", linestyle=":")
df5_pivot[('max', 'accuracy_rfc')].plot(ax=ax, label="Random forest classifier", color="red", linestyle=":")

df5_pivot[('mean', 'accuracy_welch')].plot(ax=ax, label="Welch t-test", color="blue")
df5_pivot[('min', 'accuracy_welch')].plot(ax=ax, label="Welch t-test", color="blue", linestyle=":")
df5_pivot[('max', 'accuracy_welch')].plot(ax=ax, label="Welch t-test", color="blue", linestyle=":")

ax.legend()
```

```python
def plot_metric_range(ax, xlabel, my_df_pivot, title, metric_name, metric_var, suffixes, labels, colors):
    ax.set_title(title)
    for (suffix, label, color) in zip(suffixes, labels, colors):
        my_df_pivot[('mean', f'{metric_var}_{suffix}')].plot(ax=ax, label=label, color=color)
        ax.fill_between(my_df_pivot.index,
                    my_df_pivot[('min', f'{metric_var}_{suffix}')],
                    my_df_pivot[('max', f'{metric_var}_{suffix}')],
                    alpha=0.1,
                    color=color)
    ax.set_ylabel(metric_name)
    ax.set_xlabel(xlabel)
    ax.legend()


def metric_plots_by_training_size(axes, experimental_group_size, control_group_size, tag):
    my_df = df[(df.tag == tag) & (df.control_group_size == control_group_size) & (df.experiment_group_size == experimental_group_size)]
    my_df_pivot = my_df.pivot_table(index='number_of_training_experiments', aggfunc=[min, max, 'mean', 'std', len])
    for (idx, lname, upname, varname) in [
        (0, 'accuracy', 'Accuracy', 'accuracy'),
        (1, 'type 1 error', 'Type 1 Error', 'type1'),
        (2, 'ROC/AUC', 'ROC/AUC', 'roc_auc')
    ]:
        axes[idx].set_xscale('log')
        plot_metric_range(axes[idx],
                          "Number of training experiments",
                          my_df_pivot,
                      f"Min, max and mean {lname} error\nfor generated data sets with\n{experimental_group_size} experimental and {control_group_size} control participants",
                      upname,
                      varname,
                      ['rfc', 'welch'],
                      ['Random Forest Classifier', 'Welch t-test'],
                      ['red', 'blue']
                     )
    #axes[1].set_yscale('log')
```

```python
fig, axes = matplotlib.pyplot.subplots(figsize=(20,35), nrows=6, ncols=3)
metric_plots_by_training_size(axes[0],5,5,'Participant size vs training size')
metric_plots_by_training_size(axes[1],6,6,'Participant size vs training size')
metric_plots_by_training_size(axes[2],7,7,'Participant size vs training size')
metric_plots_by_training_size(axes[3],10,10,'Participant size vs training size')
metric_plots_by_training_size(axes[4],20,20,'Participant size vs training size')
metric_plots_by_training_size(axes[5],40,40,'Participant size vs training size')
fig.tight_layout()
```

```python
my_pivot_df = df[df.number_of_training_experiments == 100000].pivot_table(
    index='control_group_size',
    aggfunc=[min, max, 'mean']
)
metric_var = 'accuracy'
suffix = 'welch'

fig, axes = matplotlib.pyplot.subplots(figsize=(20,6), ncols=3)

for (idx, lname, upname, varname) in [
        (0, 'accuracy', 'Accuracy', 'accuracy'),
        (1, 'type 1 error', 'Type 1 Error', 'type1'),
        (2, 'ROC/AUC', 'ROC/AUC', 'roc_auc')
    ]:
    plot_metric_range(axes[idx], 'Experiment size', my_pivot_df, 
                  f'{upname} by experiment size\nTraining size=100,000 experiments', 
                  upname, 
                      varname,
                   ['rfc', 'welch'],
                      ['Random Forest Classifier', 'Welch t-test'],
                      ['red', 'blue']
                 )
```
